## Exercise 1:

- You're residing in a flatshare and sharing groceries with your flatmates. Each flatmate can add items to a shared shopping list.
- Initially, create a structure with the current items: bananas, apples, and toilet paper.
- After a week, you discover that one of your lazy roommates has made another list. Merge both lists into a single one.
- You want to add soap to your shopping list.
- One of your roommates just returned from visiting their grandparents with a surplus of apples. Remove this item from your list.
- At each step, print your updated list.

**Output:**

Initial shopping list: {'bananas', 'toilet paper', 'apples'}

After merging lists: {'bread', 'milk', 'eggs', 'apples', 'bananas', 'toilet paper'}

After adding soap: {'bread', 'milk', 'eggs', 'soap', 'apples', 'bananas', 'toilet paper'}

After removing surplus apples: {'bread', 'milk', 'eggs', 'soap', 'bananas', 'toilet paper'}
